**Continuity and Differentiation**



Derivatives of functions also known in the terms of limits: f'(a) = lim(x->a) (f(x)-f(a))/(x-a). There are a few different theorems for approximating derivative functions.

Rolle's Theorem: says let f:[a,b] to R be a continuous function. And that we must assume that the derivatives exist on [a,b]. However, If f(a) = f(b), then the following is true a>c>b so that f'(c) = 0.

Mean Value Theorem: if we let f:[a,b] to R be a continous function. And we assume its derivates exist on (a,b) then the following must be true a < c < b.

Taylor Series: States that an infinite sum of terms can be expressed in the form of a function's derivatives at a single point.

The example below displays the approximation method for differential equations.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

h = 0.5
x = np.arange(0, 4*np.pi, h)
y = np.sin(x)

f_diff = np.diff(y)/h
x_diff = x[:-1:]
e_sol = np.cos(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, f_diff, '--', label = 'Finite Difference Approximation')
plt.plot(x_diff, e_sol,label = 'Exact Solution')
plt.legend()
plt.show()
```

**Unconstrained Optimization**

Some important terms to know and understand:


Global Minimizer: If we let f: R^d -> R. The point x* ∈ R^d is a global minimizer of f over R^d if f(x) ≥ f(x*) , ∀x ∈ R^d.


Local Minimizer: Let f: R^d -> R. The point x* ∈ R^d is a local minimizer of f over R^d if there is δ > 0 such that f(x) ≥ f(x) , ∀x ∈ Bδ(x) \ {x*}.


First-Order Necessary Condition: If we let f:Rd to R be continuously differentiable on Rd. If x0 is a local minimizer, then the following is true
∇f (x0)=0.


Second-Order Necessary Condition: If we let f:R^d -> R be twice continously differentiableon R^d. If x0 is the local minimizer, then H_f(x0) is PSD.


The following example demonstrates optimization.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(0,8)
y = x**2

x1 = np.linspace(0,6)
y1 = x*4.5

plt.plot(x, y,'r',x1,y1,'k-')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
