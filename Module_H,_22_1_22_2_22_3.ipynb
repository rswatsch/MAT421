**Section 22.1**

ODE Initial value Problem Statement

Differential equations are a relationship between a function f(x) and a variable x and any number of it's derivatives.

An ordinary differential equation also known as an ODE is a one dimensional differential equation with independent variables.

ode.PNG

Below is an example of a ordinary differential equation and the graph of how it looks.

```python
g=9.8
l=2
def acceleration(theta):
  return -1*g/l*np.sin(theta)

print('The acceleration of the pendulum')
theta = np.linspace(-3.14,3.14,1000)
plt.plot(theta, acceleration(theta))
plt.ylabel('d^2Theta/dt^2')
plt.xlabel('theta')
plt.show()
```

**Section 22.2, Reduction of Order**



Reduction of order is a method for solving higher order differential equations, essentially it is a method of reducing those higher order differential equations down to first order equations.

**Section 22.3, Euler Method**

Below are a few different examples of Euler method, which preforms numerical integration of the values we are interested in.

```python
dt=.1
theta0=2
maxPoints =int(10/dt)

F=np.zeros(maxPoints)
theta=theta0
dtheta = 0
t=np.zeros(maxPoints)
for i in range(maxPoints):
  atheta = acceleration(theta)
  dtheta += atheta*dt
  theta += dtheta*dt
  F[i]=theta
  t[i]=i*dt

plt.plot(t, F, 'r' ,label='Numerical Solution')
plt.plot(t, 2*np.cos(t*(g/l)**.5), label='Analytic Solution')
plt.xlabel('Time')
plt.ylabel('Theta')
plt.legend()
plt.show()
```

This next example displays the error of our approximation in comparison to the actual value.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline


f = lambda t, s: np.exp(-t)
h = 0.005
t = np.arange(0, 1 + h, h)
s0 = -1


s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 10))
plt.plot(t, s, 'k--', label='Approximate')
plt.plot(t, -np.exp(-t), 'r', label='Exact')
plt.title('Approximate and Exact Solution')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
