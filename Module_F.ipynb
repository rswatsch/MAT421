**Section 20.2**
*Finite Difference Approximating Derivatives*

In finite difference approximations there are a few different methods.


The first method is the foward difference estimate which estimates the slope of a function using the following formula which connects (x_j, f(x_j)) and (x_j+1, f(x_j+1))
f'(x_j) = (f(x_j+1) - f(x_j)) / (x_j+1 - x_j)


The second method is known as the backward difference which uses a similar connection formula to the previous as shown here, (x_j−1,f(x_j−1)) and (xj,f(xj)):

f'(x_j) = (f(x_j) - f(x_j-1)) / (x_j - x_j-1)

The final method is the central difference which utilizes these formula's. (x_j−1, f(x_j−1)) and (x_j+1, f(x_j+1)):

f'(x_j) = (f(x_j+1) - f(x_j-1)) / (x_j+1 - x_j-1)




The following example is of a foward finite difference approximation.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

# step size
h = 0.1
# define grid
x = np.arange(0, 4*np.pi, h)
# compute function
y = np.sin(x)

# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = np.cos(x_diff)

# Plot solution
plt.figure(figsize = (12, 10))
plt.plot(x_diff, forward_diff, '--', label = 'Forward Finite difference approximation')
plt.plot(x_diff, exact_solution,  label = 'Exact Solution')
plt.legend()
plt.show()


max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

This example below shows the foward difference approximation for a linearly decreasing stepsize and then plots the error of the approximated value to the true value.import

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

h = 2

iterations = 15

step_size = []

max_error = []

for i in range(iterations):
    # halve the step size
    h /= 3
    # store this step size
    step_size.append(h)
    # compute new grid
    x = np.arange(0, 2 * np.pi, h)
    # compute function value at grid
    y = np.sin(x)
    # compute vector of forward differences
    forward_diff = np.diff(y)/h
    # compute corresponding grid
    x_diff = x[:-1]
    # compute exact solution
    exact_solution = np.cos(x_diff)

    # Compute max error between
    # numerical derivative and exact solution
    max_error.append(\
            max(abs(exact_solution - forward_diff)))

# produce log-log plot of max error versus step size
plt.figure(figsize = (12, 8))
plt.loglog(step_size, max_error, 'v')
plt.show()
```

**Section 20.4**
*Numerical Differentiation with Noise*

Below is a graphical representation of how just a small amount of innaccuracy can affect when compared to a pure mathematical function.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01)
# compute function
omega = 100
epsilon = 0.01

y = np.cos(x)
y_noise = y + epsilon*np.sin(omega*x)

# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'cos(x) + noise')
plt.plot(x, y, 'k-', \
         label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
