**Simpsons Rule**

Simpsons rule approximates the area under f(x) over two subintervals that you set and by fitting a quadratic polynomial through the points which align with the interval this creates a unique polynomial that can then be integrated exactly.

**Section 21.4**

The examplem that follows is an approximation of the area under the curve of cos(x)

```python
import numpy as np

a = 0
b = np.pi
n = 20
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

print(I_simp)
print(err_simp)
```

**Section 21.5**

This next example utilizes trapz or the trapezoid rule for approximating the integral value.

```python
import numpy as np
from scipy.integrate import trapz

a = 0
b = np.pi
n = 21
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)
```

This section focuses primarily on approximating integrals using python for this next example we are utilizing cumtrapz which is an approximation of the cumulative integral of f(x)=sin(x)

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0, np.pi, 0.001)
F_exact = -np.cos(x)
F_approx = cumtrapz(np.sin(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('Integral of sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```
