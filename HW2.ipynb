# Interpolation
**What is it, how does it work and the different types**

Based on the problem statement from chapter 17.1 it is explained that we are given a data set of values consisting of xi and a corresponding yi, where i = 1,....,n.

Essentially, interpolation is a tool used to estimate a function given a certain set of data points like stated above. This is refered to as an interpolation function.

There is another similar technique called a regression something you have no doubt seen and heard of in any data or graphing class. However, interpolation does not require an underlying model for the data to be based on.

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [1,4,3,4]
y = [1,8,2,7]
plt.figure(figsize = (10,8))
plt.plot(x,y,'bo', 2.5,4,'ro')
plt.title('Interpolation data point')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Linear Interpolation 17.2
**This section will demonstrate what a linear Interpolation looks like and how it works**

Estimated points are preset to lie on the line and linear interpolation joins the nearest points to the left and the right. One trade off of this is that linear interpolation struggles to be accurate with very scattered data.

For my example, I plan to use the following data points for linear interpolation
**x = [1,2,3,6], y = [1,2,5,7]**

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')


x = [1,2,3,6]
y = [1,2,5,7]
f = interp1d(x, y)
y_hat = f(4)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(4, y_hat, 'ro')
plt.title('Linear Interpolation at x = 4')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Interpolation using Cubic Spline Technique 17.3
**In this section, I will demonstrate the Cubic Spline Interpolation style**

A cubic spline interpolation function is a set of piecewise cubic functions.


For my cubic spline interpolation example, I plan to use the following data:
**x=[0, 2, 4, 7, 8, 10], y = [1, 5, 6, 4, 5, 2]**

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 2, 4, 7, 8, 10]
y = [1, 5, 6, 4, 5, 2]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline')
plt.xlabel('x input')
plt.ylabel('y output')
plt.show()
```

# Takeaways about linear and cubic interpolation
***As we can see one of the major differences between cubic spline interpolation and linear interpolation is that the spline varation is able to bend and curve. While the linear interpolation is just a mix of straight lines from one point to another. This is the difference between the cubic polynomial function and the linear function***
